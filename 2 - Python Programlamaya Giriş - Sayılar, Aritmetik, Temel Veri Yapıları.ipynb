<h1 id="Sayılar-ve-aritmetik">Sayılar ve aritmetik<a class="anchor-link" href="#Sayılar-ve-aritmetik">¶</a></h1><p>Python'da öntanımlı (built-in) üç çeşit sayı var: <em>Tamsayılar</em>, <em>reel sayılar</em>, ve <em>karmaşık sayılar</em>. Bilimsel hesaplama yapanlar için kayan nokta temsilini iyi anlamak çok mühim, ama bunu daha sonraya bırakalım.</p>
<p>Bu üç sayı tipiyle öntanımlı olarak yedi aritmetik işlem yapılabilir: Toplama (<code>+</code>), çıkarma (<code>-</code>), çarpma (<code>*</code>), bölme (<code>/</code>), tamsayı bölme (<code>//</code>), kalan (<code>%</code>), ve üs alma (<code>**</code>).</p>


```

2 + 3

```

```

7 - 12.4

```

```

5.25 * 6.25

```

```

25 / 8

```

```

25 // 8

```

```

25 % 8

```

```

0.5 ** 1.2

```


<h2 id="Değişken-atamaları">Değişken atamaları<a class="anchor-link" href="#Değişken-atamaları">¶</a></h2><p>Python'da değişkenlerin tiplerini (int, float, vs.) önceden beyan etmek gerekmez. Değişken isimleri, atamanın sağ tarafındaki ifadeye işaret eden bir isimdir sadece.</p>


```

x = -2      # bir tamsayı
y = 3/4 * 2 # reel sayı
z = "Merhaba"+"Dünya" # dize
x,y,z

```


<p>Çoklu atamalar yapmak mümkündür:</p>


```

x, y, z = -2, 1.5, "Merhaba Dünya"
x,y,z

```


<p>Bu usulü kullanarak iki değişkenin değerini, bir ara değişken kullanmadan değiştokuş etmek mümkün olur.</p>


```

x,y = y,x
x,y

```


<h1 id="Öntanımlı-veri-yapıları">Öntanımlı veri yapıları<a class="anchor-link" href="#Öntanımlı-veri-yapıları">¶</a></h1><p>Sayılara ek olarak, Python diline dahil birkaç öntanımlı veri yapısı var. Bunları daha sonraki bölümlerde daha ayrıntılı işleyeceğiz. Şimdilik, pratik kullanıma yetecek kadar tanıtalım.</p>
<h2 id="Dizeler">Dizeler<a class="anchor-link" href="#Dizeler">¶</a></h2><p>Bir <em>dize</em> (string), karakterlerin (harf, rakam, ve diğer işaretler) birleşiminden oluşur. Tek tırnak veya çift tırnak içinde olmalıdır.</p>
<p>Dizenin içindeki karakterlere indeksleme işlemi <code>[]</code> ile erişilebilir. İndeks değerleri sıfırdan başlar; birinci karakter için 0, ikinci karakter için 1, vs. Negatif indeksler sondan başlayarak saymak için kullanılabilir: Son karakter için -1, sondan bir önceki için -2, vs. gibi.</p>


```

isim = "Albert Einstein"
isim[0] # İlk karakter

```

```

isim[1] # İkinci karakter

```

```

isim[-1] # Son karakter

```

```

isim[-2] # Sondan bir önceki karakter

```


<p>Dizeler değiştirilemez.</p>


```

z = "Merhaba"
z[0] = "m"

```


<h2 id="Listeler">Listeler<a class="anchor-link" href="#Listeler">¶</a></h2><p>Liste, köşeli parantez içinde, virgülle ayrılmış elemanları birleştiren br yapıdır. Listelerde de elemanlar farklı tiplerde olabilir. Dizeler için geçerli olan eleman indeksleme işlemleri listelerde de aynen geçerlidir. Listelerin farkı <em>değiştirilebilir</em> (mutable) olmalarıdır, yani elemanlarına tekrar atama yapılabilir.</p>


```

liste = [5, "merhaba", (2,3,1), [-1,0,2]]
liste[0] = 4.1  # birinci elemanı değiştir
liste[-1][0] = 12 # son elemanın birinci elemanını değiştir
liste

```

```

liste[1] = "naber" # İkinci elemana başka bir dize ata.
liste

```

```

del liste[2] # Üçüncü elemanı sil
liste

```


<p>Ancak, listenin bir elemanı değiştirilemez bir tipteyse, o elemanın alt elemanlarını silemez veya onlara atama yapamazsınız.</p>


```

liste[1][0] = "h"

```


<h2 id="Sözlükler">Sözlükler<a class="anchor-link" href="#Sözlükler">¶</a></h2><p>Dize, çokuz, ve liste tipleri <em>sıralı tip</em> (sequence type) olarak anılır, çünkü birinci elemandan son elemana kadar iyi tanımlanmış bir düzenleri mevcuttur. Sözlük (dictionary) tipi ise sıralı değildir. Sözlükler, listelerin genelleştirilmiş halidir: Listelerde elemanların referanslarının sıfırdan başlayan tamsayılar olması gerekirken, sözlüklerde ise çok değişik veri tipleri her şey referans olarak kullanılabilir. Bu tür veri tipleri <em>hash table</em>, <em>hash map</em>, veya <em>associative array</em> olarak da bilinirler.</p>


```

d = {-1.75: "merhaba", "isim":"Einstein"}
d[-1.75]

```

```

d["isim"]

```

```

d[(2,3)] = [1,2,3.14159]
d

```

```

d[-2.25] = 3 + 5j
d[1+2j] = 42
d["isim"] = "Eisenstein"
d

```

```

del d["isim"]
d

```


<p>Sözlük yaratırken süslü parantezler <code>{}</code> kullanırız. Sözlük elemanlarına ise, çokuzlar ve listelerdeki gibi, köşeli parantez içinde referans değerini yazarak erişiriz. Yeni referans değerlerine sahip sözlük elementleri yaratabilir, istemediğimiz elemanları <code>del</code> komutuyla silebiliriz.</p>
<p>Sözlüklerde referans olarak sadece <em>değiştirilemez</em> tipler kullanılabilir (sayılar, dizeler, çokuzlar). Listeler kullanılamaz.</p>
