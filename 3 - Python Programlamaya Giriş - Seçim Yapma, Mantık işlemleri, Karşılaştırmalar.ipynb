
<p>Yeni başlayanlar için Python programlama temellerini işlediğimiz yazı dizimizin bu bölümünde karar verme yapısından bahsedeceğiz.</p>
<p>En basit programlarda bile sık sık "şu şart doğruysa şöyle yap, yoksa yapma", veya "doğruysa şöyle yap, yanlışsa öbür türlü yap" şeklinde düzenlemelere ihtiyaç duyarız. Sözgelişi, "cevap doğruysa puanı bir arttır", veya "yanlış sayısı dördün katıysa puanı bir azalt" gibi. Birçok dil gibi Python'da da seçim yapısı <code>if-else</code> komutlarıyla kurulur.</p>
<h1 id="if-komutu">if komutu<a class="anchor-link" href="#if-komutu">¶</a></h1><p>Karar yapılarının en basit hali, bir şartın doğruluğunu yoklamak ve doğruysa belli işlemler yapmaktır. Sözgelişi aşağıdaki programa bakalım. Hücreyi Shift-Enter ile çalıştırdığımızda tahminimizi soran bir kutu ortaya çıkacak. Kutuya bir cevap yazıp Enter'e basın.</p>


```

hedef = "42"
tahmin = input("Tahmininiz: ")
if tahmin == hedef:
    print("Doğru!")

```


<p>Sayı <code>hedef</code> değerine (<code>"42"</code>) eşitse ekrana <code>"Doğru!"</code> yazısını basacak, değilse birşey yapmayacak. Şarttan sonra iki nokta üstüste (<code>:</code>) koymayı da unutmayın.</p>
<p>(Bu hücredeki kodu kendi başına çalışan bir betik haline getirmek için, kodu kopyalayıp bir metin editörüne yapıştırın ve tahmin.py isimli bir dosya olarak kaydedin. Böylece komut satırında çalıştırabilirsiniz:</p>
<pre><code>$ python3 tahmin.py
Tahmininiz: 42
Doğru!
</code></pre>
<p>şeklinde bir çıktınız olmalı)</p>
<p><code>if</code>'den sonra <code>tahmin</code>'in <code>hedef</code>'e eşit olup olmadığını yoklayan bir mantık ifadesi var. Bunun değeri ya doğru (<code>True</code>) ya da yanlış (<code>False</code>) olacak. Eşitlik şartı için <code>==</code> (iki tane eşit işareti) kullanıldığına dikkat edin.</p>


```

1 == 1

```

```

1 == 2

```

```

1 = 1

```


<p>Şarttan sonraki satırlara, şart doğruysa çalıştırılacak kod parçası (<em>if bloku</em>) gelir. Burada Python'un kendine özgü bir kuralını görüyoruz: Bloktaki her ifade, <code>if</code> başlığına göre belli bir miktar sağa kaydırılmalıdır. Yorumlayıcı hangi komutların  <code>if</code> blokuna ait olduğunu bu kaydırma sayesinde ayırt eder.</p>

<pre><code>hedef = "42"
tahmin = input("Tahmininiz: ")

    print("Doğru!")
    print("Bravo")
</code></pre>

tahmin = input("Tahmininiz: ")
if tahmin == hedef:
    print("Doğru!")
print("Bravo")

<p>komutu <code>if</code> blokunun dışında olduğu için, tahmin doğru olsa da olmasa da çalıştırılır. Deneyin.</p>
<p><code>if</code> içindeki blokun kaç boşluk sağa kaydırılacağı önemli değildir; bir tek boşluk bile yeterlidir. Çoğu IDE otomatik olarak dört boşluk genişlikte bir sıçrama yapar; Python programcılık camiasında tavsiye edilen de budur. Sonraki satırlar da aynı kaydırma seviyesinde başlar. Bu yüzden, bloku bitirmek için yeni satıra geçtikten sonra geriye silme (Backspace) tuşuna basmalısınız.</p>
<h1 id="if-else-komutu">if-else komutu<a class="anchor-link" href="#if-else-komutu">¶</a></h1><p>Yukarıdaki program bize sadece tahminimiz doğruysa bir cevap veriyor. Tahminimizin yanlış olduğunu da söylemesini istersek <code>if-else</code> yapısını kullanırız.</p>


```

hedef = "42"
x = input("Tahmininiz: ")
if x == hedef:
    print("Doğru!")
else:
    print("Yanlış")

```


<p><code>else</code> ifadesinden sonra herhangi bir şart gelmez, çünkü <code>else</code> bloku zaten sadece <code>if</code> şartı yanlışsa çalıştırılır. <code>else</code> kelimesinin, bağlı olduğu <code>if</code> ile aynı kaydırma seviyesinde olduğuna dikkat edin.</p>
<p>Her <code>else</code> bir <code>if</code>'e bağlı olmalıdır, ama tersi doğru değildir. Her <code>if</code>'e bir <code>else</code> gerekmez.</p>
<h1 id="if-elif-else-komutu">if-elif-else komutu<a class="anchor-link" href="#if-elif-else-komutu">¶</a></h1><p><code>if</code> kullanımının en genel hali budur. <code>elif</code> kelimesi <code>else if</code>'in kısaltmasıdır.</p>


```

x = int(input("Kaç tane? "))
if x>2000:
    print("Binlerce")
elif x>200:
    print("Yüzlerce")
elif x>10:
    print("Çok")
elif x>0:
    print("Birkaç")
else:
    print("Yok")

```


<p>Aynı programı <code>elif</code> kullanmadan ancak şöyle yazabiliriz:</p>


```

x = int(input("Kaç tane? "))
if x>2000:
    print("Binlerce")
else:
    if x>200:
        print("Yüzlerce")
    else:
        if x>10:
            print("Çok")
        else:
            if x>0:
                print("Birkaç")
            else:
                print("Yok")

```


<p>Görüldüğü gibi <code>elif</code> kullanmak programı sadeleştiriyor.</p>
<h1 id="Karşılaştırmalar">Karşılaştırmalar<a class="anchor-link" href="#Karşılaştırmalar">¶</a></h1><p><code>if</code>'den sonra, mantıksal olarak doğru veya yanlış sayılacak ifadeler gelmelidir. Eşitlikten yukarıda bahsettik. Ayrıca <code>&gt;</code> (büyük), <code>&lt;</code> (küçük), <code>&gt;=</code> (büyük veya eşit), <code>&lt;=</code> (küçük veya eşit) ve <code>!=</code> (eşit değil) işlemleri kullanılabilir.</p>


```

2 > 3, 2 >= 3, 2 < 3, 2 <= 3, 2 != 3

```


<p>Sayı olmayan veri yapıları da karşılaştırmalarda kullanılabilirler. O durumda, eşitsizliklerin değerlendirilmesinde alfabetik sıra kullanılır.</p>


```

"hello" == "Hello", "hello" > "Hello", "hello" < "jello"

```

```

[1,2,3] < [1,20,3], [1,2,3] < [11,2,3]

```


<p>Daha fazla bilgi için <a href="http://docs.python.org/2/reference/expressions.html#not-in">Python dili referansına</a> bakabilirsiniz.</p>
<p>Bir veri yapısı içinde belli bir eleman veya alt grubun bulunup bulunmadığını <code>in</code> kelimesiyle test edebilirsiniz. Şartı ters çevirmek için <code>not in</code> kullanılır.</p>


```

L = [12,3,4,[5,6]]
3 in L, [3,4] in L, [5,6] in L, [5,6] not in L

```

```

s = "merhaba"
"a" in s, "erh" in s

```


<p>Sözlüklerde <code>in</code> kelimesi sadece referanslar içinde yoklama yapar.</p>


```

d = {"abc": 54, (1,2): -45.1}
"abc" in d, 54 in d, (1,2) in d

```


<h1 id="Doğru-ve-yanlış-sabitleri">Doğru ve yanlış sabitleri<a class="anchor-link" href="#Doğru-ve-yanlış-sabitleri">¶</a></h1><p><code>True</code> ve <code>False</code>, aslında 1 ve 0 sayılarına verilen yeni isimlerdir. Ayrıca,</p>
<pre><code>[], {}, "", None, 0, 0.0, False
</code></pre>
<p>ifadelerinin her biri mantıksal yanlış anlamı taşır. Tersine olarak da, sıfırdan farklı her sayı veya boş olmayan herhangi bir liste/çokuz/dize/sözlük <code>if</code> yapılarında mantıksal doğru olarak yorumlanırlar.</p>


```

L=[]
if L:
    print("Liste dolu")
else:
    print("Liste boş.")

```

```

s = "abc"
if s:
    print(s)
else:
    print("Boş")

```


<h1 id="Mantıksal-işlemler">Mantıksal işlemler<a class="anchor-link" href="#Mantıksal-işlemler">¶</a></h1><p>Birden fazla şartı Boole işlemleri (<code>and</code>, <code>or</code>, <code>not</code>) ile biraraya getirerek, daha karmaşık şartlar oluşturmak mümkün olur.</p>
<ul>
<li><code>X and Y</code> : Hem <code>X</code> hem <code>Y</code> doğruysa doğru; <code>X</code> ve <code>Y</code>'den en az biri yanlışsa yanlış.</li>
<li><code>X or Y</code> : <code>X</code> ve <code>Y</code>'den en az biri doğruysa doğru; hem <code>X</code> hem <code>Y</code> yanlışsa yanlış.</li>
<li><code>not X</code> : <code>X</code>'in doğruluk değerinin tersi.</li>
</ul>


```

1 < 2 and 3==3

```

```

1 < 2 and 3==4

```

```

1 < 2 or 3==4

```

```

1 > 2 or 3==4

```

```

not 1>2

```

```

not 3==3

```


<p>Boole ifadelerinde ikiden fazla bileşen de kullanılabilir. Önce <code>not</code> işlemi, sonra <code>and</code> işlemi, sonra da <code>or</code> işlemi yapılır. Aynı işlem yapılıyorsa, bileşenler soldan sağa çifter çifter alınır.</p>


```

1 == 2 or 3 > 4 or 5 < 8

```

```

1 != 2 and not 3 > 4 and 5 < 8

```

```

1 != 2 and 3 <= 4 or 5 > 10 and 7 > 8

```


<h1 id="Üçlü-if-else-ifadesi">Üçlü if-else ifadesi<a class="anchor-link" href="#Üçlü-if-else-ifadesi">¶</a></h1><p>Birçok durumda, bir değişkene, bir şartın doğru veya yanlış olmasına göre farklı değerler atarız. Sözgelişi, <code>x</code> mantık ifadesinin doğru olması halinde <code>a</code>'ya <code>y</code> değeri verelim, aksi takdirde <code>z</code> verelim.</p>
<pre><code>if x:
    a = y
else:
    a = z
</code></pre>
<p>Üçlü <code>if-else</code> ifadesiyle aynısını daha kısa olarak şöyle yazabiliriz:</p>
<pre><code>a = y if x else z
</code></pre>
<p>Örnek olarak, <code>m</code>'ye <code>x</code>'in mutlak değerini atayalım.</p>


```

x = float(input("Bir sayı girin: "))
m = x if x>0 else -x
print(m)

```


<p>En baştaki örneği de şöyle yazabiliriz.</p>


```

hedef = "42"
x = input("Tahmininiz: ")
print("Doğru!" if x == hedef else "Yanlış")

```


<p>Bir <code>if-elif-else</code> blokunu üçlü <code>if-else</code> ifadesi olarak yazabiliriz. Sözgelişi</p>
<pre><code>if rating &gt; 100:
    sinif = "A"
elif rating &gt; 50:
    sinif = "B"
else:
    sinif = "C"
</code></pre>
<p>yerine</p>
<pre><code>sinif = "A" if rating &gt; 100 else "B" if rating &gt; 50 else "C"
</code></pre>
<p>yazılabilir. Ama aşırıya kaçırırsanız programın okunması zorlaşır.</p>
